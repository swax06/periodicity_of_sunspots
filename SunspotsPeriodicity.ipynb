```python
import numpy as np
import pandas as pd
from scipy.signal import detrend
from plotly.offline import iplot, init_notebook_mode
import plotly.graph_objs as go
```

```python
df = pd.read_csv("SN_ms_tot_V2.0.csv", delimiter=';')
x = np.array(df["Year"])
y = np.array(df["Mean value"])
trace = go.Scatter(x = x, y = y,mode= "lines", line= dict(shape = "spline"))
data = [trace]
layout = {"title": "Monthly number of Sunspots since 1749", "xaxis_title":"Years", "yaxis_title":"No. of Sunspots"}
iplot({"data":data, "layout":layout})
```

```python
sampling_frequency = 12 #in years
sampling_interval = 1/sampling_frequency
```

```python
#converting to ac
y = detrend(y)
trace = go.Scatter(x = x, y = y,mode= "lines", line= dict(shape = "spline"))
data = [trace]
layout = {"title": "Monthly number of Sunspots since 1749", "xaxis_title":"Years", "yaxis_title":"No. of Sunspots"}
iplot({"data":data, "layout":layout})
```

```python
fourierTransform = np.fft.fft(y)/len(y)        
#removing the period of input data repeatation
fourierTransform = fourierTransform[range(int(len(y)/2))] 
samples     = len(y)
values      = np.arange(int(samples/2))
timePeriod  = samples/sampling_frequency
frequencies = values/timePeriod
```

```python
trace = go.Scatter(x = frequencies, y = abs(fourierTransform),mode= "lines", line= dict(shape = "spline"))
data = [trace]
layout = {"title": "Fourier transform of no. of sunspots", "xaxis_title":"Frequency", "yaxis_title":"Value"}
iplot({"data":data, "layout":layout})
```

```python
peak_location = np.argmax(abs(fourierTransform))
peak_value = frequencies[peak_location]
print("1st frequency: ", 1/peak_value,"years")
print("2nd frequency : ", 1/0.118, "years")
```
